```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
df = pd.read_csv("../csvs/batch_results.csv")
```

```python
df.head(15)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='maxContourArea', hue='gesture', palette='Set1', bins=40, kde=True)
plt.title('Distribution of Max Contour Area by Gesture')
plt.xlabel('Max Contour Area')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='gesture', hue='convexDefects', palette='Set1', bins=40, kde=True)
plt.title('Distribution of Max Contour Area by Gesture')
plt.xlabel('Max Contour Area')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
sns.lmplot(
    data=df, x="maxContourArea", y="avgAngle", col="gesture", hue="fingers",
    col_wrap=2, palette="muted", ci=None,
    height=4, scatter_kws={"s": 50, "alpha": 1}
)
```

```python

plt.figure(figsize=(6, 6))
df['gesture'].value_counts().plot.pie(autopct='%.0f%%', startangle=00, colors=sns.color_palette('Set3'))
plt.ylabel('')
plt.title('Distribuio de Gestos (Pizza)')
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='gesture', palette='Set2')
plt.title('Distribuio de Gestos')
plt.xlabel('Gestos')
plt.ylabel('Contagem')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

plt.figure(figsize=(6, 6))
df['fingers'].value_counts().plot.pie(autopct='%.0f%%', startangle=00, colors=sns.color_palette('Set3'))
plt.ylabel('')
plt.title('Distribuio de Dedos (Pizza)')
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='fingers', palette='Set2')
plt.title('Distribuio de Dedos')
plt.xlabel('Dedos')
plt.ylabel('Contagem')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Histograma de 'fingers'
plt.figure(figsize=(8, 5))
sns.histplot(df['fingers'], bins=range(df['fingers'].min(), df['fingers'].max() + 2), color='skyblue', kde=False)
plt.title('Distribuio de Dedos Detectados')
plt.xlabel('Quantidade de Dedos')
plt.ylabel('Nmero de Imagens')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Boxplot de avgAngle por fingers
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='fingers', y='avgAngle', palette='Set2')
plt.title('Variao do ngulo Mdio por Quantidade de Dedos')
plt.xlabel('Quantidade de Dedos')
plt.ylabel('ngulo Mdio')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Histograma de maxContourArea
plt.figure(figsize=(8, 5))
sns.histplot(df['maxContourArea'], bins=40, color='mediumseagreen', kde=True)
plt.title('Distribuio da rea Mxima do Contorno da Mo')
plt.xlabel('maxContourArea')
plt.ylabel('Nmero de Imagens')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Scatter plot: maxContourArea vs convexDefects
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='maxContourArea', y='convexDefects', hue='gesture', palette='Set1', alpha=0.7)
plt.title('Relao entre rea Mxima do Contorno e Defeitos Convexos')
plt.xlabel('maxContourArea')
plt.ylabel('convexDefects')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Gesture')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='centerX', y='centerY', hue='gesture', palette='Set1', alpha=0.6)
plt.title('Mapa de Posies da Mo nas Imagens (Scatter)')
plt.xlabel('centerX')
plt.ylabel('centerY')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Gesture')
plt.show()

plt.figure(figsize=(8, 6))
sns.kdeplot(
    data=df, x='centerX', y='centerY', fill=True, cmap='viridis', thresh=0.05, levels=100
)
plt.title('Mapa de Calor das Posies da Mo nas Imagens (Heatmap)')
plt.xlabel('centerX')
plt.ylabel('centerY')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Bar chart: frequncia de cada gesto nas imagens estticas
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='gesture', palette='Set2')
plt.title('Frequncia de Cada Gesto nas Imagens Estticas')
plt.xlabel('Gesto')
plt.ylabel('Contagem')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Boxplot: maxContourArea por gesture
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='gesture', y='maxContourArea', palette='Set3')
plt.title('Distribuio da rea Mxima do Contorno por Gesto')
plt.xlabel('Gesto')
plt.ylabel('maxContourArea')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Scatter: fingers vs maxContourArea
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='maxContourArea', y='fingers', hue='gesture', palette='Set1', alpha=0.7)
plt.title('Fingers vs Max Contour Area')
plt.xlabel('maxContourArea')
plt.ylabel('fingers')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend(title='Gesture')
plt.show()

# Scatter: convexDefects vs avgAngle por gesto
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='convexDefects', y='avgAngle', hue='gesture', palette='Set2', alpha=0.7)
plt.title('Convex Defects vs Avg Angle por Gesto')
plt.xlabel('convexDefects')
plt.ylabel('avgAngle')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend(title='Gesture')
plt.show()

# Heatmap de centerX, centerY por gesture
g = sns.FacetGrid(df, col="gesture", col_wrap=3, height=4)
g.map_dataframe(sns.kdeplot, x="centerX", y="centerY", fill=True, cmap='viridis', thresh=0.05, levels=100)
g.set_titles(col_template="{col_name}")
g.set_axis_labels("centerX", "centerY")
for ax in g.axes.flatten():
    ax.grid(True, linestyle='--', alpha=0.7)
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Heatmap de centerX, centerY por Gesto')
plt.show()
```
